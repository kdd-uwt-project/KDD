```python
distance_array = []
```

```python
eta_array = []
```

```python
transportMode_array = []
```

```python
for sIndex in train_plan:
    for priceRecord in sIndex[2]:
        if(priceRecord["price"] == ''):
            distance_array.append(priceRecord["distance"])
            eta_array.append(priceRecord["eta"])
            transportMode_array.append(priceRecord["transport_mode"])
```

```python
distance_dic = {}
eta_dic = {}
transportMode_dic = {}
```

```python
for dist in distance_array:
    if(dist not in distance_dic):
        distance_dic[dist] = 1
    else:
        distance_dic[dist] += 1
```

```python
for etas in eta_array:
    if(etas not in eta_dic):
        eta_dic[etas] = 1
    else:
        eta_dic[etas] += 1
```

```python
for trans in transportMode_array:
    if(trans not in transportMode_dic):
        transportMode_dic[trans] = 1
    else:
        transportMode_dic[trans] += 1
```
